2. Musim Panas (Summer):
Musim panas memiliki rata-rata temperatur yang tinggi, tetapi sedikit lebih rendah dari musim gugur. Kecepatan angin rata-rata lebih tinggi dari musim gugur, menunjukkan kondisi angin yang sedikit lebih kencang.

3. Musim Gugur (Fall):
Musim gugur memiliki rata-rata temperatur (temp dan atemp) tertinggi, menunjukkan kondisi cuaca yang paling hangat. Kecepatan angin rata-rata terendah, menunjukkan kondisi angin yang paling tenang.

4. Musim Dingin (Winter):
Musim dingin memiliki rata-rata temperatur terendah, menunjukkan kondisi cuaca yang paling dingin. Kecepatan angin rata-rata rendah, tetapi sedikit lebih tinggi dari musim gugur


## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh kondisi cuaca dan musim terhadap pola penyewaan sepeda

```python
# Visualisasi distribusi penyewaan per musim
seasonal_avg = hour_df.groupby('season', observed=True)['count'].mean()

# Membuat barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=seasonal_avg.index, y=seasonal_avg.values)
plt.title('Rata-rata Penyewaan Sepeda per Musim')
plt.xlabel('Musim')
plt.ylabel('Rata-rata Jumlah Penyewaan')

# Menampilkan plot
plt.show()
```

```python
correlation = hour_df[['temp', 'atemp', 'hum', 'windspeed', 'count']].corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Korelasi antara Variabel Cuaca dan Jumlah Penyewaan')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='count', hue='weathersit', data=hour_df)
plt.title('Tren Penyewaan Sepeda per Kondisi Cuaca')
plt.xlabel('Waktu')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

**Insight:**
- Secara keseluruhan, tren penyewaan sepeda tampak meningkat dari waktu ke waktu, terutama pada cuaca cerah dan kabut.
Ini menunjukkan peningkatan minat terhadap penyewaan sepeda selama periode waktu yang ditampilkan.
- Hujan lebat/es memiliki jumlah penyewaan yang sangat rendah. Ini menunjukkan bahwa orang sangat menghindari bersepeda dalam kondisi cuaca ekstrem.
- Ada pola musiman yang terlihat, dengan puncak penyewaan terjadi pada bulan mei sampai bulan juli dan penurunan setelah bulan oktober sampai dengan januari

### Pertanyaan 2: Bagaimana pola penyewaan sepeda bervariasi berdasarkan hari dalam seminggu dan jam dalam sehari?

```python
# Agregasi data
daily_counts = hour_df.groupby('weekday', observed=True)['count'].sum()

# Visualisasi dengan barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_counts.index, y=daily_counts.values)
plt.title('Jumlah Penyewaan Sepeda per Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu)')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

```python
# Hitung penyewaan per jam
hourly_counts = hour_df.groupby('hour')['count'].sum()

#visualisasi puncak penyewaan sepeda berdasarkan jam
plt.figure(figsize=(12, 6))
plt.bar(hourly_counts.index, hourly_counts.values)
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.title('Distribusi Penyewaan Sepeda per Jam')
plt.xticks(range(24))
plt.show()
```

```python
# Agregasi jumlah penyewaan harian
daily_counts = hour_df.groupby('date')['count'].sum()

# Agregasi jumlah penyewaan per jam per hari
daily_hourly_avg = hour_df.groupby(['date', 'hour'])['count'].mean().groupby('date').mean()

# Hitung korelasi
correlation = daily_counts.corr(daily_hourly_avg)
print(f"Korelasi antara jumlah penyewaan harian dan rata-rata penyewaan per jam per hari: {correlation}")

# Visualisasi with scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(daily_counts, daily_hourly_avg)
plt.title('Scatter Plot: Jumlah Penyewaan Harian vs. Rata-rata Penyewaan Per Jam Per Hari')
plt.xlabel('Jumlah Penyewaan Harian')
plt.ylabel('Rata-rata Penyewaan Per Jam Per Hari')
plt.show()
```

Pola titik-titik pada scatter plot menunjukkan korelasi positif yang sangat kuat. Artinya, ketika jumlah penyewaan harian meningkat, rata-rata penyewaan per jam per hari juga meningkat secara signifikan. Titik-titik data hampir membentuk garis lurus, yang mengindikasikan hubungan linear yang kuat antara kedua variabel.

```python
# Grafik Titik 
g = sns.FacetGrid(hour_df, row='weekday', height=2, aspect=4)
g.map(sns.pointplot, 'hour', 'count')
g.set_axis_labels('Jam', 'Jumlah Penyewaan')
plt.show()
```
